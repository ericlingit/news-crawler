```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
```

```python
data = Path('./spiders/output/four_categories/wantwant')
data.exists()
```

### Store in a DataFrame

Load all raw data

```python
raw = pd.DataFrame()
```

```python
for fi in (data/'raw').iterdir():
    with open(fi) as f:
        raw = raw.append(json.load(f), ignore_index=True)
```

```python
raw
```

```python
raw.to_json(data/'wantwant_raw.json')
```

---

### Clean text

#### Remove html elements

```python
raw.title[:5].str.normalize('NFKC').str.replace('<[^>]*>', '').str.strip()
```

```python
raw.title[:2].str.replace('<[^>]*>', '').str.normalize('NFKC')[0]
```

```python
clean = pd.DataFrame()
```

```python
clean['category'] = raw.category
clean['date'] = raw.date
```

```python
clean['title'] = raw.title.str.replace('<[^>]*>', '').str.normalize('NFKC')
```

```python
clean['text'] = raw.text.str.replace('<[^>]*>', '').str.normalize('NFKC')
```

```python
clean['url'] = raw.url
```

```python
clean
```

#### Change category labels:

- change `star` to `entertainment`
- change `money` to `finance`
- change `sports` to `sport`

```python
clean['category'] == 'star'
```

```python
ent = clean[clean['category'] == 'star']
ent
```

```python
ent = ent.reset_index()
```

```python
ent['category'] = 'entertainment'
ent
```

```python
ent
```

```python
fin = clean[clean['category'] == 'money']
fin.shape
```

```python
fin['category'] = 'finance'
```

```python
spo = clean[clean['category'] == 'sports']
spo.shape
```

```python
spo['category'] = 'sport'
```

```python
clean2 = clean2.append(fin, ignore_index=True)
clean2 = clean2.append(spo, ignore_index=True)
clean2 = clean2.append(ent, ignore_index=True)
```
