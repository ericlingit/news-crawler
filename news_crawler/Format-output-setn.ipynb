```python
import json
import pandas as pd
from pathlib import Path
```

```python
src = Path('./output/news_categorical/setn')
```

```python
x = pd.DataFrame()

for fi in src.iterdir():
    with open(fi) as f:
        j = json.load(f)
    x = x.append(j, ignore_index=True)
```
